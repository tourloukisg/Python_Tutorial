```python
# UniversityStudent class example --> Use of a method(function) that takes 'self' as argument

class UniversityStudent:

    def __init__(self,name,surname,tuition):
        self.name=name
        self.surname=surname
        self.tuition=tuition
        self.e_address=name+'_'+surname+'@uni.com'

# The studentinfo method takes 'self' as an argument and returns the name,surname, email and tuition for each
# object(student)
    def studentinfo(self):
        return 'Student name: {} {} - Student Email: {} - Tuition:{}'.format(self.name,self.surname,self.e_address,self.tuition)


student_1=UniversityStudent('George','Smith',16000)
student_2=UniversityStudent('Tom','Jones',20000)

print('First Student Info:')
print(student_1.studentinfo())
print('\r')
print('Second Student Info:')
print(student_2.studentinfo())
```

```python
# UniversityStudent class example --> Use of a method(function) and a class variable

class UniversityStudent:

# 'tuition_perc_increase' is a class variable that is available to all class instances/objects. 
# Class variables are used within classes but they are not defined inside the methods(functions)
    tuition_perc_increase=0.05

    def __init__(self,name,surname,tuition):
        self.name=name
        self.surname=surname
        self.tuition=tuition
        self.e_address=name+'_'+surname+'@uni.com'

# The 'new_tuition' class method takes the  '__init__ 'argument (tuition) and updates its value 
# The class variable 'tuition_perc_increase' can be used for any class instance ('self.tuition_perc_increase')
    def new_tuition(self):
        self.tuition=self.tuition *(1+self.tuition_perc_increase)


student_1=UniversityStudent('George','Smith',16000)
student_2=UniversityStudent('Tom','Jones',20000)

print('Current Tuition Fees-First Student:')
print(student_1.tuition)
print('\r')
print('Current Tuition Fees-Second Student:')
print(student_2.tuition)
print('----------------------------')
# Calling the method for each class object/instance
student_1.new_tuition()
student_2.new_tuition()
print('New Tuition Fees-First Student:')
print(student_1.tuition)
print('\r')
print('New Tuition Fees-Second Student:')
print(student_2.tuition)
```

```python
# UniversityStudent class example --> Use of @classmethod decorator

class UniversityStudent:
    tuition_perc_increase=0.05

    def __init__(self,name,surname,tuition):
        self.name=name
        self.surname=surname
        self.tuition=tuition
        self.e_address=name+'_'+surname+'@uni.com'

    def new_tuition(self):
        self.tuition=self.tuition *(1+self.tuition_perc_increase)

# The role of the 'classmethod' decorator is to indicate that this method is not an object method and that
# it is related to the class. That is why it does not take the instance(self) as an argument. When the
# class method is used, the first argument that takes as input is usually termed as 'cls' (from class).
    @classmethod
    def change_tuition_perc_increase(cls,percentage):
        cls.tuition_perc_increase=percentage

student_1=UniversityStudent('George','Smith',16000)
student_2=UniversityStudent('Tom','Jones',20000)

print('Current Tuition Fees-First Student:')
print(student_1.tuition)
print('\r')
print('Current Tuition Fees-Second Student:')
print(student_2.tuition)

# In this case, the class method is called using the name of the class 'UniversityStudent' but it can also
# be called using a class object/instance
UniversityStudent.change_tuition_perc_increase(0.09)
print('\r')
print('---------------------------------')
print('Tuition fees (%) increase:',UniversityStudent.tuition_perc_increase)
student_1.new_tuition()
student_2.new_tuition()
print('---------------------------------')

print('New Tuition Fees-First Student:')
print(student_1.tuition)
print('\r')
print('New Tuition Fees-Second Student:')
print(student_2.tuition)
```

```python
# UniversityStudent class example --> Use of class variables inside the '__init__' method and
# Use of '@classmethod' as an alternative constructor

class UniversityStudent:

    # class variable that counts the number of students (initially set to 0)
    num_students=0

    def __init__(self,name,surname,tuition):
        self.name=name
        self.surname=surname
        self.tuition=tuition
        self.e_address=name+'_'+surname+'@uni.com'

# Use of class variable inside the '__init__' method (value increases by one each time a new object is created)
        UniversityStudent.num_students+=1

# In this example, the @classmethod is being used to create a method that takes as inputs the 'cls' argument
# and an entry of string type where the object's attributes are seperated by '_'. This class method separates
# each argument with the use of split() and returns a class instance (student_3 info)
    @classmethod
    def studentconstructor(cls,new_entry):
        name,surname,tuition=new_entry.split('_')
        return cls(name,surname,tuition)

student_1=UniversityStudent('George','Smith',16000)
student_2=UniversityStudent('Tom','Jones',20000)

# New student entry in different format
new_student="Anna_Harris_22000"

# Calling the studentconstructor method using the name of the class. It takes as input argument the new entry.
student_3=UniversityStudent.studentconstructor(new_student)

print('Current Tuition Fees-First Student:')
print(student_1.__dict__)
print('\r')
print('Current Tuition Fees-Second Student:')
print(student_2.__dict__)
print('\r')
print('Current Tuition Fees-Third Student:')
print(student_3.__dict__)
print('\r')
print('Total number of Students -->',UniversityStudent.num_students)
```

```python
# UniversityStudent class example --> Use of @staticmethod decorator
# Instance methods use the instance as the first argument (self)
# Class methods use the class as the first argument(cls)
# Static method does not use any of the aforementioned arguments as first input within a function.
# The method input consists of arguments that are meaningful to the created static method  and
# logically related to the class (University Students).

class UniversityStudent:

    def __init__(self,name,surname,tuition):
        self.name=name
        self.surname=surname
        self.tuition=tuition
        self.e_address=name+'_'+surname+'@uni.com'


# In this static method, the argument that is passed as input is 'date'. This method(function) returns
# whether a student is on summer break or not(if month= 7 'July' or if month=8 'August', then on break).
# As it can be observed, there is no use of instance (self) or class (cls) arguments.
    @staticmethod
    def isinsummerbreak(date):
        if date.month==7 or date.month==8:
            return True
        return False

student_1=UniversityStudent('George','Smith',16000)
student_2=UniversityStudent('Tom','Jones',20000)
       
# Importing datetime and selecting a date
import datetime
date=datetime.date(2021,8,20)

# Calling the static method using the class name 'UniversityStudent'.
print(UniversityStudent.isinsummerbreak(date))  

# Calling the static method using the class instance/object.
print(student_1.isinsummerbreak(date))  
```

```python
# UniversityStudent class example --> Use of __init__ method inside the sub-class for initialization purposes

class UniversityStudent:

    tuition_perc_increase=0.05

    def __init__(self,name,surname,tuition):
        self.name=name
        self.surname=surname
        self.tuition=tuition
        self.e_address=name+'_'+surname+'@uni.com'

    def new_tuition(self):
        self.tuition=self.tuition *(1+self.tuition_perc_increase)

# lawstudent class inherits all methods and attributes from the UniversityStudent class
class lawstudent(UniversityStudent):

# Use of __init__ methos inside the lawstudent class to add a new argument.
# There is use of super().__init__ to forward the first three arguments to the UniversityStudent class
    def __init__(self,name,surname,tuition,academic_year):
        super().__init__(name,surname,tuition)
        self.academic_year=academic_year

lawstudent_1=lawstudent('George','Smith',16000,'1st')
lawstudent_2=lawstudent('Tom','Jones',20000,'2nd')

print(lawstudent_1.academic_year)
print(lawstudent_2.academic_year)
```

```python
# UniversityStudent class example --> Use of i)multiple sub-classes, ii) isinstance(), iii) issubclass()

class UniversityStudent:

    def __init__(self,name,surname):
        self.name=name
        self.surname=surname
        self.e_address=name+'_'+surname+'@uni.com'

    def new_tuition(self):
        self.tuition=self.tuition *(1+self.tuition_perc_increase)

class lawstudent(UniversityStudent):

    def __init__(self,name,surname,academic_year):
        super().__init__(name,surname)
        self.academic_year=academic_year

# Creating a new sub-class lawprofessor. A new argument is passed as input to the __init__ method and 
# it indicates the number of student this professor supervises.
class lawprofessor(UniversityStudent):

    def __init__(self,name,surname,law_student=None):
        super().__init__(name,surname)

# If the professor supervises no students return an empty list, else return the students
        if law_student==None:
            self.law_student=[]
        else:
            self.law_student=law_student

# Function to append a new student that is to be supervised by the same professor
    def supervise_new_student(self,student):
        if student not in self.law_student:
            self.law_student.append(student)

# Function to display all students that are supervised by the same professor    
    def display_student_info(self):
        for student in self.law_student:
            print(student.__dict__)

lawstudent_1=lawstudent('George','Smith','1st')
lawstudent_2=lawstudent('Tom','Jones','2nd')
lawstudent_3=lawstudent('Paul','Brown','3nd')

law_prof=lawprofessor('Sam','West',[lawstudent_1,lawstudent_2])

print('Law Professor e-address:\n',law_prof.e_address)
print('-------------------------------------------------------------------------------------------------------')
print('Students supervised by the Professor:\n')
law_prof.display_student_info()
print('-------------------------------------------------------------------------------------------------------')

# Appending a new law student
law_prof.supervise_new_student(lawstudent_3)

print('Students supervised by the Professor-New Entry:\n')
law_prof.display_student_info()
print('-------------------------------------------------------------------------------------------------------')

# Use of isinstance()--> Checking whether an object is an istance of a class or not
print('Is lawstudent_1 an instant of the lawstudent class? -->',isinstance(lawstudent_1,lawstudent))
print('Is lawstudent_1 an instant of the UniversityStudent class? -->',isinstance(lawstudent_1,UniversityStudent))
print('Is lawstudent_1 an instant of the lawprofessor class? -->',isinstance(lawstudent_1,lawprofessor))
print('Is law_prof an instant of the lawprofessor class? -->',isinstance(law_prof,lawprofessor))
print('-------------------------------------------------------------------------------------------------------')

# Use of issubclass()--> Checking whether an class is a subclass of another class or not
print('Is lawstudent a subclass of UniversityStudent class ? -->',issubclass(lawstudent,UniversityStudent))
print('Is lawprofessor a subclass of UniversityStudent class ? -->',issubclass(lawprofessor,UniversityStudent))
print('Is lawprofessor a subclass of lawstudent class ? -->',issubclass(lawprofessor,lawstudent))
```
